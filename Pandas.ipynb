**Lambda Function Column and Row in DataFrame**

```python
# Split name for students into first and last name
First_name = lambda x: x.split()[0]
students['First_name'] = students.Name.apply(First_name)
```

```python
students
```

```python
# Split name for students into first and last name
Last_name = lambda x: x.split()[-1]
students['Last_name'] = students.Name.apply(Last_name)
```

```python
students['Blood'] = students.Blood_Type.apply(
    lambda x: x.split('-')[0]
    )
```

```python
students
```

```python
students['Blood'] = students.Blood_Type.apply(
    lambda x: x.split('-')[-1]
    )
```

```python
students
```

```python
# Applying Lambda to a row
# Checking if students is tall to fly
students['can_play'] = students.apply(lambda row:
     'yes'
     if row['Height'] >= 4.5
     else 'No',
     axis=1
)
```

```python
students
```

```python
# if house is gryffindor then brave 
students['brave'] = students.House.apply(lambda x: 'Yes' if x =='Gryffindor' else 'No')
```

```python
students
```

```python
# Adding Safe in First_name if Pure-Blood
students['pure_safe'] = students.apply(lambda row : 'Safe ' + row['First_name']
                                       if row['Blood_Type']=='Pure-Blood'
                                      else 'Not_safe ' + row['First_name'],axis=1)
```

```python
students
```

```python
# removing extra colums
students.drop(['First_name', 'Last_name','Blood','brave','pure_safe'], axis = 1)
```
